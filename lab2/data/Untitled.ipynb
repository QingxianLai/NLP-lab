```python
import numpy as np
import theano
```

```python
f = open("imdb.pkl", 'rb')
```

```python
train = cPickle.load(f)
```

```python
trainX_sample = train[0][:20]
```

```python
lengths = [len(x) for x in trainX_sample]
print lengths
```

```python
n_sample = len(trainX_sample)
maxlen = np.max(lengths)
```

```python
x = np.zeros((maxlen, n_sample)).astype('int64')
```

```python
x_mask = np.zeros((maxlen, n_sample)).astype(theano.config.floatX)
```

```python
for idx, s in enumerate(trainX_sample):
    x[:lengths[idx], idx] = s
    x_mask[:lengths[idx], idx] = 1.
```
